```python
import pandas as pd
```

```python
### 读取文件，默认参数
iris_data = pd.read_csv("../_Datasets/iris.data")  #sep指定字段分隔符，默认逗号。
iris_data.head()
```

```python
### sep
#读取文件，指定字段分隔符
iris_data = pd.read_csv("../_Datasets/iris.data",sep=',')  
iris_data.head()
```

```python
### encoding
#读取文件，指定编码
#如果出现中文乱码，可更换为中文编码"gb2312"或"gbk"
iris_data = pd.read_csv("../_Datasets/iris.data", encoding="utf-8")  
iris_data.head()
```

```python
### header
## 默认header,默认取第0行作为header
# 注意对比iris.data和iris_noheader.data2个原始数据的差异
iris_data2 = pd.read_csv("../_Datasets/iris_noheader.data") 
iris_data2.head()
```

```python
## 指定header
#header参数可以是一个list例如：[0,1,3]，这个list表示将文件中的这些行作为列标题（意味着每一列有多个标题），介于中间的行将被忽略掉
#注意：真实数据很多时候没有那么规范，且存在多级标题的可能性。
iris_data3 = pd.read_csv("../_Datasets/iris_noheader.data",header=4)   #取第1行作为header，上面的行抛弃
iris_data3.head()
```

```python
#作为header的行，不再作为数据行。以下可知iris_data2的数据少了1行,iris_data3的数据少了5行
print(iris_data.shape)
print(iris_data2.shape)
print(iris_data3.shape)
```

```python
## 不指定header，默认给1个列名
iris_data4 = pd.read_csv("../_Datasets/iris_noheader.data",header=None)  
iris_data4.head()
```

```python
iris_data4.shape
```

```python
### names 
#指定列名
#iris_noheader.data没有字段名，默认第0行是字段名，但实际第0行是数据行，这样有问题，需要设置字段名。
#例如把字段名设置为[sepal_length,sepal_width,petal_length,petal_width,species]
iris_data5 = pd.read_csv("../_Datasets/iris_noheader.data",
                         names=["sepal_length","sepal_width","petal_length","petal_width","species"])  
iris_data5.head()
```

```python
### dtypes 
#指定数据格式， dtype是字典结构
iris_data6 = pd.read_csv("../_Datasets/iris_noheader.data",
                         names=["sepal_length","sepal_width","petal_length","petal_width","species"],
                         dtype={"sepal_length":float})  
iris_data6.head()
```

```python
### nrows
#设置读取行数，对大文件可部分读取。
iris_data7 = pd.read_csv("../_Datasets/iris.data",nrows=50)  
iris_data7.head()
```

```python
iris_data7.shape
```

```python
### skiprows 
#从头部跳过多少行。注意先过滤，然后再确定表头
iris_data8 = pd.read_csv("../_Datasets/iris.data",skiprows=50)  
iris_data8.head()
```

```python
iris_data8.shape
```

```python
### skipfooter 
#从文件末尾过滤行
iris_data9 = pd.read_csv("../_Datasets/iris.data",skipfooter=50)  
iris_data9.head()
```

```python
iris_data9.shape
```

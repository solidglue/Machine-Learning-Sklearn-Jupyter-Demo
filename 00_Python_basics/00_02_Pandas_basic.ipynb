```python
## Series
# Series： 类似于一维数组或列表，是由一组数据以及与之相关的数据标签（索引）构成。
#Series 可以看作是 DataFrame 中的一列，也可以是单独存在的一维数据结构。

ser1 = iris_data["species"]
print(ser1,'\n')  #Serises
print(type(ser1)) # 查看ser1类型
```

```python
# Series转DataFrame
df1 = pd.DataFrame(ser1)
print(df1.head(),'\n')
print(type(df1))  # 查看df1类型
```

```python
### 0.2.1 concat拼接dataframe 

##按列concat

#首先按列拆分DataFrame为多个DafaFrame，再合并
#例如拆分前几列为训练集,选多个列时要用二维数组表示。
train_data = iris_data[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
train_data.head()
```

```python
# 最后一列拆分成标签。
# 注意iris_data["species"]返回的是Series，iris_data[["species"]]返回的是DataFrame。
train_label = iris_data[["species"]]
train_label.head()
```

```python
type(train_label)
```

```python
# 按列合并特征数据和标签
# axis：默认是0代表按行合并，如果等于1代表按列合并
iris_data1 = pd.concat([train_data,train_label],axis=1) 
iris_data1
```

```python
## 按行concat

# 按行拆分数据集 - 前70行拆分为df1，70行之后拆分成df2
df1 = iris_data.iloc[:70,]
df1
```

```python
df2 = iris_data.iloc[70:,]
df2
```

```python
#axis：默认是0代表按行合并，如果等于1代表按列合并
iris_data2 = pd.concat([df1,df2],axis=0)
iris_data2
```

```python
# reset index
# 看看按列concat df1和df2会怎样
df3 = pd.concat([df1,df2],axis=1)
df3
```

```python
# 按行concat是根据index拼接的。df1的index是0-70，df2的index是80-149
# 以上操作相当于sql里边的外连接。如果想df1与df2按数据对齐合并，需要重置df2的index

df2_ = df2.reset_index(drop=True)
df2_ #此时df2_的索引被重置成了0-80
```

```python
#再拼接df1和df2_
df3_ = pd.concat([df1,df2_],axis=1)
df3_   #此时前70行对其了，后10行因为df1没有数据，因此显示NAN。
```

```python
#如果我们想要df1和df2共同的index数据，我们可以指定join方式，类似sql里的内连接、左链接和右链接。
df4 = pd.concat([df1,df2_],join='inner',axis=1)
df4  #此时只保留了都有的index数据,与df3_对比
```

```python
### 0.2.3 merge拼接dataframe

# 以上可知，concat方法可根据index join 2个DataFrame,如果我们想按照某个列join则无法实现。

#merge，通过merge实现sql里的join操作。on表示连接字段。
df6 = pd.merge(df1,df2,on=['sepal_length','sepal_width'],how='inner')
df6  
```

```python
### 0.2.4 join拼接dataframe
# 除concat join和merge外，join方法也可以实现sql里的表连接，主要按index join.
df7 = df1.join(df2_,how='inner',lsuffix='_left',rsuffix='_right',on='sepal_length')
df7  #join字段重名冲突，需要手动设置，而merge会自动重命名。并且多字段join会有奇奇怪怪的错误需要处理。
```

```python
### 0.2.5 排序
# 按 sepal_length进行排序
iris_data3 = iris_data.sort_values(by="sepal_length", ascending=True)
iris_data3
```

```python
### 0.2.6 删除某列
df8 = iris_data3.drop(["petal_length","petal_width"],axis=1)
df8
```

```python
### 0.2.7 算术运算 & 增加列
# 例如计算sepal_length和sepal_width差值，构建新特征列sepal_div
df8["sepal_div"] = df8["sepal_length"] - df8["sepal_width"] 
df8
```

```python
### 小结

#merge方法主要是通过key将frame合并，这个key可以是DataFrame中的列（columns）也可以是索引（index）。
#而join方法主要是通过index进行合并（也支持列名，但用起来比较繁琐），它默认使用index-on-index的方式进行合并。
#同理，concat的join也只支持index合并。

# 推荐使用merge。
```
